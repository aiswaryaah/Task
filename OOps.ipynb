#### Encapsulation
* hiding and protecting data
* putting data and methods together in one object,and controlling who can access or change that data

```python
class BankAccount:
    def __init__(self,owner,initial_bal):
        self.owner=owner                             
        self.__balance=initial_bal       #double' _' means its private
    def deposit(self,amount):
        if amount<=0:
            print("Deposit amount should be greater then zero")
            return          #return is to store
        self.__balance=self.__balance+amount
        print(f"₹{amount} deposited successfully.")
        print(f"New balance:₹{self.__balance}")
    def withdraw(self,amount):
        if amount<=0:
            print("ERROR:Withdrawal amount must be greater than zero")
            return
        if amount>self.__balance:
            print("ERROR:Insufficient balance!")       
            return
        self.__balance=self.__balance-amount
        print(f"₹{amount} withdrwan successfully.")
        print(f"Remaining balance: ₹{self.__balance}")
    def get_balance(self):
        print(f"{self.owner}'s current balance:₹{self.__balance}")
joel_account=BankAccount("Joel",10000)
joel_account.deposit(5000)
joel_account.withdraw(2000)
joel_account.get_balance()
print()
print("---What happens when you try to break the rules---")
joel_account.withdraw(50000)
joel_account.deposit(-100)
print()
print("---What happens if you try to directly access private data---")
```

### Inheritance
* Parents class is the base class
* Child class is the derived class
  

```python
class Employee:

    def __init__(self, name, emp_id, salary):
        self.name = name          # Every employee has a name
        self.emp_id = emp_id      # Every employee has an ID
        self.salary = salary      # Every employee has a salary

    def clock_in(self):
        # Every employee clocks in — defined once, available to all child classes
        print(f"{self.name} (ID: {self.emp_id}) has clocked IN.")

    def clock_out(self):
        # Every employee clocks out
        print(f"{self.name} (ID: {self.emp_id}) has clocked OUT.")

    def show_details(self):
        print(f"Name    : {self.name}")
        print(f"Emp ID  : {self.emp_id}")
        print(f"Salary  : Rs.{self.salary}/month")


# ---- CHILD CLASS 1 — Manager ----
class Manager(Employee):
    # The (Employee) here means: Manager INHERITS from Employee
    # Manager automatically gets: __init__, clock_in, clock_out, show_details

    def __init__(self, name, emp_id, salary, team_size):
        super().__init__(name, emp_id, salary)
        # super() means: "go to the parent class (Employee) and call its __init__"
        # This sets self.name, self.emp_id, self.salary — without rewriting that code

        self.team_size = team_size
        # This is something ONLY a Manager has — not in Employee

    def assign_task(self, task, employee_name):
        # This is a NEW method — only Manager has this, Employee doesn't
        print(f"Manager {self.name} assigned task '{task}' to {employee_name}.")

    def show_details(self):
        # We are OVERRIDING the parent's show_details to add extra info
        super().show_details()             # First call the parent's version
        print(f"Role    : Manager")
        print(f"Team    : {self.team_size} members")


# ---- CHILD CLASS 2 — Intern ----
class Intern(Employee):

    def __init__(self, name, emp_id, salary, college):
        super().__init__(name, emp_id, salary)
        # Note: we pass stipend as salary — it still uses Employee's __init__

        self.college = college
        # Only interns have a college name

    def show_details(self):
        super().show_details()
        print(f"Role    : Intern")
        print(f"College : {self.college}")
        print(f"(Note: Salary shown above is monthly stipend)")


# ---- Creating Objects ----

print("===== MANAGER =====")
mgr = Manager("Rahul Sharma", "MGR001", 80000, 10)
mgr.clock_in()              # Inherited from Employee — no need to rewrite
mgr.show_details()          # Overridden version — shows extra info
mgr.assign_task("Build login page", "Joel")  # Manager-only method
mgr.clock_out()             # Inherited from Employee

print()
print("===== INTERN =====")
intern1 = Intern("Joel", "INT022", 15000, "SRM IST")
intern1.clock_in()          # Inherited from Employee
intern1.show_details()      # Overridden version
intern1.clock_out()         # Inherited from Employee
```

### Polymorphism
* poly means many
* morph means froms
  

```python
#Polymorphism



# ---- PARENT CLASS ----
class Notification:

    def __init__(self, recipient):
        self.recipient = recipient
        # Every notification has a recipient (person to notify)

    def send(self, message):
        # This is a BASE version — child classes will override this
        print(f"Sending notification to {self.recipient}: {message}")


# ---- CHILD CLASS 1 — Email ----
class EmailNotification(Notification):

    def __init__(self, recipient, email_address):
        super().__init__(recipient)
        self.email_address = email_address

    def send(self, message):
        # OVERRIDING the parent's send() with email-specific logic
        print(f"[EMAIL] Connecting to SMTP server...")
        print(f"[EMAIL] To      : {self.email_address}")
        print(f"[EMAIL] Message : {message}")
        print(f"[EMAIL] Email sent successfully to {self.recipient}!")


# ---- CHILD CLASS 2 — SMS ----
class SMSNotification(Notification):

    def __init__(self, recipient, phone_number):
        super().__init__(recipient)
        self.phone_number = phone_number

    def send(self, message):
        # Same method name — completely different behavior
        print(f"[SMS] Connecting to telecom gateway...")
        print(f"[SMS] To      : +91 {self.phone_number}")
        print(f"[SMS] Message : {message}")
        print(f"[SMS] SMS delivered to {self.recipient}!")


# ---- CHILD CLASS 3 — WhatsApp ----
class WhatsAppNotification(Notification):

    def __init__(self, recipient, phone_number):
        super().__init__(recipient)
        self.phone_number = phone_number

    def send(self, message):
        # Same method name — WhatsApp-specific behavior
        print(f"[WHATSAPP] Opening WhatsApp Business API...")
        print(f"[WHATSAPP] To      : +91 {self.phone_number}")
        print(f"[WHATSAPP] Message : {message}")
        print(f"[WHATSAPP] Message delivered on WhatsApp to {self.recipient}!")


# ---- The POWER of Polymorphism ----
# This function doesn't know or care WHICH type of notification it gets
# It just calls .send() — and each object handles it its own way

def notify_customer(notification_object, message):
    notification_object.send(message)
    # That's it. One line. Works for Email, SMS, WhatsApp — anything.
    print()


# Create different notification objects
email_notif    = EmailNotification("Joel", "joel@example.com")
sms_notif      = SMSNotification("Arjun", "9876543210")
whatsapp_notif = WhatsAppNotification("Priya", "8765432109")

order_message = "Your order #FL5523 has been shipped! Expected delivery: Tomorrow."

# Calling the SAME function with DIFFERENT objects
# Same method name .send() — totally different behavior each time
notify_customer(email_notif,    order_message)
notify_customer(sms_notif,      order_message)
notify_customer(whatsapp_notif, order_message)
```

```python
'''Task 1 — Encapsulation 
Problem 
Build a MobilePhone class. 
A mobile phone stores a PIN. Nobody should be able to directly read or set the PIN from outside the 
class. You can only: 
• Set PIN through set_pin(new_pin) — but only if the new PIN is exactly 4 digits (a number 
between 1000 and 9999) 
• Verify PIN through check_pin(entered_pin) — returns True if correct, False if wrong 
• The actual PIN should never be printed anywhere outside the class 
Try to run this: 
phone = MobilePhone("Samsung Galaxy", 4821) 
phone.check_pin(1234)        
phone.check_pin(4821)        
phone.set_pin(99)            
phone.set_pin(7654)          
phone.check_pin(4821)        
phone.check_pin(7654)        
Expected Output: 
Samsung Galaxy initialized. 
# Wrong PIN 
# Correct PIN 
# Should reject — not 4 digits 
# Should accept — valid 4 digit PIN 
# Should now be wrong (PIN changed) 
# Should be correct now 
Incorrect PIN. Access denied. 
Correct PIN. Phone unlocked! 
Invalid PIN. Must be a 4-digit number (1000-9999). 
PIN updated successfully. 
Incorrect PIN. Access denied. 
Correct PIN. Phone unlocked!'''

class MobilePhone:
    def __init__(self, model, pin):
        self.model = model
        self.__pin = None
        if 1000 <= pin <= 9999:
            self.__pin = pin
        print(f"{self.model} initialized.")
    def set_pin(self, new_pin):
        if 1000 <= new_pin <= 9999:
            self.__pin = new_pin
            print("PIN updated successfully.")
        else:
            print("Invalid PIN. Must be a 4-digit number (1000-9999).")
    def check_pin(self, entered_pin):
        if entered_pin == self.__pin:
            print("Correct PIN. Phone unlocked!")
        else:
            print("Incorrect PIN. Access denied.")
phone = MobilePhone("Samsung Galaxy", 4821)
phone.check_pin(1234)
phone.check_pin(4821)
phone.set_pin(99)
phone.set_pin(7654)
phone.check_pin(4821)
phone.check_pin(7654)
```

```python
'''Task 2 — Inheritance 
Problem 
You are building a school system. Create: 
1. A parent class Person with: 
o name and age attributes 
o A method introduce() that prints: "Hi, I am [name] and I am [age] years old." 
2. A child class Student that inherits from Person with: 
o An extra attribute student_id 
o An extra attribute grade (like "A", "B", etc.) 
o Override introduce() to also print their student ID and grade 
3. A child class Teacher that inherits from Person with: 
o An extra attribute subject (the subject they teach) 
o Override introduce() to also print their subject 
o An extra method take_class(topic) that prints: "[name] is teaching [topic] today." 
Try to run this: 
s = Student("Arun", 20, "S2045", "A") 
t = Teacher("Mrs. Kavitha", 35, "Mathematics") 
s.introduce() 
print() 
t.introduce() 
t.take_class("Quadratic Equations") 
Expected Output: 
Hi, I am Arun and I am 20 years old. 
Student ID : S2045 
Grade      
: A 
Hi, I am Mrs. Kavitha and I am 35 years old. 
Subject    : Mathematics 
Mrs. Kavitha is teaching Quadratic Equations today.'''

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def introduce(self):
        print(f"Hi, I am {self.name} and I am {self.age} years old.")
class Student(Person):
    def __init__(self, name, age, student_id, grade):
        super().__init__(name, age)
        self.student_id = student_id
        self.grade = grade
    def introduce(self):
        super().introduce()
        print(f"Student ID : {self.student_id}")
        print(f"Grade      : {self.grade}")
class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject
    def introduce(self):
        super().introduce()
        print(f"Subject    : {self.subject}")
    def take_class(self, topic):
        print(f"{self.name} is teaching {topic} today.")
s = Student("Arun", 20, "S2045", "A")
t = Teacher("Mrs. Kavitha", 35, "Mathematics")
s.introduce()
print()
t.introduce()
t.take_class("Quadratic Equations")
```
